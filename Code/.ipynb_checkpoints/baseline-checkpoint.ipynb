# Model

## Baseline에서 제공하는 Model 사용

```python
def make_oof_prediction(train, y, test, features, categorical_features='auto', model_params=None, folds=10,model_name = 'lightgbm'):
    x_train = train[features]
    x_test = test[features]
    
    # 테스트 데이터 예측값을 저장할 변수
    test_preds = np.zeros(x_test.shape[0])
    
    # Out Of Fold Validation 예측 데이터를 저장할 변수
    y_oof = np.zeros(x_train.shape[0])
    
    # 폴드별 평균 Validation 스코어를 저장할 변수
    score = 0
    
    # 피처 중요도를 저장할 데이터 프레임 선언
    fi = pd.DataFrame()
    fi['feature'] = features
    
    # Stratified K Fold 선언
    skf = StratifiedKFold(n_splits=folds, shuffle=True, random_state=SEED)

    for fold, (tr_idx, val_idx) in enumerate(skf.split(x_train, y)):
        # train index, validation index로 train 데이터를 나눔
        x_tr, x_val = x_train.loc[tr_idx, features], x_train.loc[val_idx, features]
        y_tr, y_val = y[tr_idx], y[val_idx]
        
        print(f'fold: {fold+1}, x_tr.shape: {x_tr.shape}, x_val.shape: {x_val.shape}')
        
        # LightGBM 
        if model_name =='lightgbm':
            dtrain = lgb.Dataset(x_tr, label=y_tr)
            dvalid = lgb.Dataset(x_val, label=y_val)
        
        # LightGBM 모델 훈련
            clf = lgb.train(
                model_params,
                dtrain,
                valid_sets=[dtrain, dvalid], # Validation 성능을 측정할 수 있도록 설정
                categorical_feature=categorical_features,
                verbose_eval=200,    
            )

            # Validation 데이터 예측
            val_preds = clf.predict(x_val)
            
        elif model_name =='catboost':
            clf = CatBoostClassifier(**model_params)
            clf.fit(x_tr, y_tr,
                    eval_set=(x_val, y_val), # Validation 성능을 측정할 수 있도록 설정
                    cat_features=categorical_features,
                    use_best_model=True,
                    verbose=True)

            # Validation 데이터 예측
            val_preds = clf.predict_proba(x_val)[:,1]
            
        elif model_name =='xgb':
            dtrain = xgb.DMatrix(x_tr, label=y_tr)
            dvalid = xgb.DMatrix(x_val, label=y_val)

            # XGBoost 모델 훈련
            clf = xgb.train(
                model_params,
                dtrain,
                num_boost_round=10000, # 트리 개수
                evals=[(dtrain, 'train'), (dvalid, 'valid')],  # Validation 성능을 측정할 수 있도록 설정
                verbose_eval=200,
                early_stopping_rounds=100
            )

            # Validation 데이터 예측
            val_preds = clf.predict(dvalid)    
        
        # Validation index에 예측값 저장 
        y_oof[val_idx] = val_preds        
                
        # 폴드별 Validation 스코어 측정
        print(f"Fold {fold + 1} | AUC: {roc_auc_score(y_val, val_preds)}")
        print('-'*80)

        # score 변수에 폴드별 평균 Validation 스코어 저장
        score += roc_auc_score(y_val, val_preds) / folds
        
        # 테스트 데이터 예측하고 평균해서 저장
        test_preds += clf.predict(x_test) / folds
        
        # 폴드별 피처 중요도 저장
        fi[f'fold_{fold+1}'] = clf.feature_importance()

        del x_tr, x_val, y_tr, y_val
        gc.collect()
        
    print(f"\nMean AUC = {score}") # 폴드별 Validation 스코어 출력
    print(f"OOF AUC = {roc_auc_score(y, y_oof)}") # Out Of Fold Validation 스코어 출력    
    
        
    # 폴드별 피처 중요도 평균값 계산해서 저장 
    fi_cols = [col for col in fi.columns if 'fold_' in col]
    fi['importance'] = fi[fi_cols].mean(axis=1)
    
#     test_preds = test_preds.round(0).astype(int)
    return y_oof, test_preds, fi
```

# feature_engineering

```python
def generate_label(df, year_month, total_thres=300, print_log=False):
    df = df.copy()
    
    # year_month에 해당하는 label 데이터 생성    
    df['year_month'] = df['order_date'].dt.strftime('%Y-%m')
    df.reset_index(drop=True, inplace=True)

    # year_month 이전 월의 고객 ID 추출
#     cust = df[df['year_month']<year_month]['customer_id'].unique()
    cust = df['customer_id'].unique()
    # year_month에 해당하는 데이터 선택
    df = df[df['year_month']==year_month]
    
    # label 데이터프레임 생성
    label = pd.DataFrame({'customer_id':cust})
    label['year_month'] = year_month
    
    # year_month에 해당하는 고객 ID의 구매액의 합 계산
    grped = df.groupby(['customer_id','year_month'], as_index=False)[['total']].sum()
    
    # label 데이터프레임과 merge하고 구매액 임계값을 넘었는지 여부로 label 생성
    label = label.merge(grped, on=['customer_id','year_month'], how='left')
    label['total'].fillna(0.0, inplace=True)
    label['label'] = (label['total'] > total_thres).astype(int)

    # 고객 ID로 정렬
    label = label.sort_values('customer_id').reset_index(drop=True)
    if print_log: print(f'{year_month} - final label shape: {label.shape}')
    
    return label
```

```python
#base line 사용
#결측치와 범주형 데이터 처리
def feature_preprocess(train, test, features, do_imputing=True):
    x_tr = train.copy()
    x_te = test.copy()
    
    # 범주형 피처 이름을 저장할 변수
    cate_cols = []

    # 레이블 인코딩
    for f in features:
        if x_tr[f].dtype.name == 'object': # 데이터 타입이 object(str)이면 레이블 인코딩
            cate_cols.append(f)
            le = LabelEncoder()
            # train + test 데이터를 합쳐서 레이블 인코딩 함수에 fit
            le.fit(list(x_tr[f].values) + list(x_te[f].values))
            
            # train 데이터 레이블 인코딩 변환 수행
            x_tr[f] = le.transform(list(x_tr[f].values))
            
            # test 데이터 레이블 인코딩 변환 수행
            x_te[f] = le.transform(list(x_te[f].values))

    print('categorical feature:', cate_cols)
    

    if do_imputing:
        # 중위값으로 결측치 채우기
#         imputer = SimpleImputer(strategy='median')
        imputer = SimpleImputer(strategy='constant', fill_value=0)

        x_tr[features] = imputer.fit_transform(x_tr[features])
        x_te[features] = imputer.transform(x_te[features])
    
    return x_tr, x_te
```

# Col 추가하기

```python
def cumsum_customer_id_total(df):
    df['cumsum_total'] = df.groupby(['customer_id'])['total'].cumsum()
    return df['cumsum_total']

def cumsum_order_id_total(df):
    df['cumsum_order_id'] = df.groupby(['order_id'])['total'].cumsum()
    return df['cumsum_order_id']    

def last_bought(df):
    df['year_month'] = df['order_date'].dt.strftime('%Y-%m')
    data = df.groupby(['customer_id','year_month'],as_index=False).sum()
    data = data.drop_duplicates('customer_id', keep='last')
    data['last_month'] = pd.to_datetime(data['year_month']).dt.month
    return data[['customer_id','last_month']]
    
def first_bought(df):
    df['year_month'] = df['order_date'].dt.strftime('%Y-%m')
    data = df.groupby(['customer_id','year_month'],as_index=False).sum()
    data = data.drop_duplicates('customer_id', keep='first')
    data['first_month'] = pd.to_datetime(data['year_month']).dt.month    
    return data[['customer_id','first_month']]    

def time_diff(df):
    df['order_ts'] = df['order_date'].astype(np.int64)//1e9
#     customer_by_order_ts_diff = df.groupby(['customer_id','order_date'])['order_ts'].mean().reset_index()
#     df['order_ts_diff'] = customer_by_order_ts_diff.groupby(['customer_id'])['order_ts'].diff()
    df['order_ts_diff'] = df.groupby(['customer_id'])['order_ts'].diff()
#     df.drop(['order_ts'],axis=1,inplace=True)
    return df['order_ts_diff']

def time_sum(df):
    df['order_ts'] = df['order_date'].astype(np.int64)//1e9
    df['order_ts_sum'] = df.groupby(['customer_id'])['order_ts'].sum()
    return df['order_ts_sum']
```

```python
def feature_engineering(df, train_start,train_end,test_start,test_end):
    df = df.copy()          
   
    # train, test 레이블 데이터 생성
    train_label = generate_label(df, train_end)[['customer_id','year_month','label']]
    test_label = generate_label(df, test_end)[['customer_id','year_month','label']]
    
    #표준화 작업
#     std_scaler = MinMaxScaler() #StandardScaler
#     df['total'] = std_scaler.fit_transform(df[['total']])
    
    # train, test 데이터 선택
    train = df[(df['order_date'] < train_end) & (df['order_date'] > train_start) ]  
    test = df[(df['order_date'] < test_end) & (df['order_date'] > test_start) ] 
    
    #train col 추가    
    train['cumsum_total'] = cumsum_customer_id_total(train)    
    train['cumsum_order_id'] = cumsum_order_id_total(train)    
    train['order_ts_diff'] = time_diff(train)    
#     train['order_ts_sum'] = time_sum(train)    
    
    #test col 추가
    test['cumsum_total'] = cumsum_customer_id_total(test)    
    test['cumsum_order_id'] = cumsum_order_id_total(test)
    test['order_ts_diff'] = time_diff(test)
#     test['order_ts_sum'] = time_sum(test)
       
    
    #description nunique 추가
    cols = ['order_id','description','product_id'] #description product_id order_id
    train_agg_nuique = train.groupby(['customer_id'])[cols].agg(['nunique'])
    test_agg_nuique = test.groupby(['customer_id'])[cols].agg(['nunique'])
        
    data_cols = []
    for col in train_agg_nuique.columns.levels[0]:
            for stat in train_agg_nuique.columns.levels[1]:
                data_cols.append(f'{col}-{stat}')
                
    train_agg_nuique.columns = data_cols
    train_agg_nuique.reset_index(inplace = True)
    test_agg_nuique.columns = data_cols 
    
    
    #first month
    train_first_month = first_bought(train)
    test_first_month = first_bought(test)
    
    #Last month
    train_last_month = last_bought(train)
    test_last_month = last_bought(test)
            
    
    # lambda 추가, 가장많이 구매한 달 , 구매한 달 목록의 길이   
    train['month'] = train['order_date'].dt.month
    test['month'] = test['order_date'].dt.month        
    train_month = train.groupby(['customer_id'])['month'].agg([lambda x : x.value_counts().index.tolist()[0],lambda x : len(x.value_counts().index.tolist())])
    test_month = test.groupby(['customer_id'])['month'].agg([lambda x : x.value_counts().index.tolist()[0],lambda x : len(x.value_counts().index.tolist())])
    train_month.rename(columns = {'<lambda_0>' : 'month_count_max','<lambda_1>' : 'month_count_len'}, inplace = True)
    test_month.rename(columns = {'<lambda_0>' : 'month_count_max','<lambda_1>' : 'month_count_len'}, inplace = True)        
    train.drop(['month'],axis=1,inplace=True)
    test.drop(['month'],axis=1,inplace=True)    
    
    
   
    # group by aggregation 함수 선언
#     agg_func = ['mean','max','min','sum','count','std','skew']
    agg_func = ['mean','sum','std','skew'] # 'count'
    all_train_data = pd.DataFrame()    
    train_agg = train.groupby(['customer_id']).agg(agg_func)

    # 멀티 레벨 컬럼을 사용하기 쉽게 1 레벨 컬럼명으로 변경
    new_cols = []
    for col in train_agg.columns.levels[0]:
        for stat in train_agg.columns.levels[1]:
            new_cols.append(f'{col}-{stat}')

    train_agg.columns = new_cols
    train_agg.reset_index(inplace = True)
    all_train_data = all_train_data.append(train_agg)        
    
     # group by aggretation 함수로 test 데이터 피처 생성
    test_agg = test.groupby(['customer_id']).agg(agg_func)
    test_agg.columns = new_cols   
    
   
    
    #data merge , 추가한 feature
    all_train_data = train_label.merge(all_train_data, on=['customer_id'], how='left')    
    all_train_data = all_train_data.merge(train_agg_nuique,on =['customer_id'],how = 'left')
    all_train_data = all_train_data.merge(train_month,on =['customer_id'],how = 'left')
    all_train_data = all_train_data.merge(train_last_month,on =['customer_id'],how = 'left')
    all_train_data = all_train_data.merge(train_first_month,on =['customer_id'],how = 'left')
    
    
    #제거한 feature
#     all_train_data['month_total_mean'] = all_train_data['total-sum']/all_train_data['month_count_len']
#     all_train_data['month_total_count_mean'] = all_train_data['total-count']/all_train_data['month_count_len']
#     all_train_data['month_cumsum_total_mean'] = all_train_data['cumsum_total-sum']/all_train_data['month_count_len']
#     all_train_data['month_diff_f_l'] = all_train_data['first_month']-all_train_data['last_month']
#     all_train_data['month_diff_l_f'] = all_train_data['last_month']-all_train_data['first_month']
    
    
    test_data = test_label.merge(test_agg, on=['customer_id'], how='left')
    test_data = test_data.merge(test_agg_nuique,on =['customer_id'],how = 'left')
    test_data = test_data.merge(test_month,on =['customer_id'],how = 'left')
    test_data = test_data.merge(test_last_month,on =['customer_id'],how = 'left')
    test_data = test_data.merge(test_first_month,on =['customer_id'],how = 'left')
    
    
#     test_data['month_total_mean'] = test_data['total-sum']/test_data['month_count_len']
#     test_data['month_total_count_mean'] = test_data['total-count']/test_data['month_count_len']
#     test_data['month_cumsum_total_mean'] = test_data['cumsum_total-sum']/test_data['month_count_len']
#     test_data['month_diff_f_l'] = test_data['first_month']-test_data['last_month']
#     test_data['month_diff_l_f'] = test_data['last_month']-test_data['first_month']
    
    #범주형 데이터로 바꾸기
#     all_train_data['first_month'] = all_train_data['first_month'].astype(str)
#     all_train_data['last_month'] = all_train_data['last_month'].astype(str)
#     test_data['first_month'] = test_data['first_month'].astype(str)
#     test_data['last_month'] = test_data['last_month'].astype(str)       
    
    
    features = all_train_data.drop(columns=['customer_id', 'label','year_month']).columns #'year_month'  
          
    # train, test 데이터 전처리
    x_tr, x_te = feature_preprocess(all_train_data, test_data, features)
    
    print('x_tr.shape', x_tr.shape, ', x_te.shape', x_te.shape)
    
    return x_tr, x_te, all_train_data['label'], features
```

```python
train_label = generate_label(data, '2010-12')['label']
```

```python
#lightGBM
def objective(trial, label=train_label,model_name = model_name):
    if model_name == 'lightgbm':
        model_params = {
            'objective': 'binary', # 이진 분류
            'boosting_type': 'gbdt',
            'num_leaves': trial.suggest_int('num_leaves', 2, 256), # num_leaves 값을 2-256까지 정수값 중에 사용
            'max_bin': trial.suggest_int('max_bin', 128, 256), # max_bin 값을 128-256까지 정수값 중에 사용
            # min_data_in_leaf 값을 10-40까지 정수값 중에 사용
            'min_data_in_leaf': trial.suggest_int('min_data_in_leaf', 10, 40),
            # 피처 샘플링 비율을 0.4-1.0까지 중에 uniform 분포로 사용
            'feature_fraction': trial.suggest_uniform('feature_fraction', 0.4, 1.0),
            # 데이터 샘플링 비율을 0.4-1.0까지 중에 uniform 분포로 사용
            'bagging_fraction': trial.suggest_uniform('bagging_fraction', 0.4, 1.0),
            # 데이터 샘플링 횟수를 1-7까지 정수값 중에 사용
            'bagging_freq': trial.suggest_int('bagging_freq', 1, 7),
            'n_estimators': 10000, # 트리 개수
            'early_stopping_rounds': 100,
            # L1 값을 1e-8-10.0까지 로그 uniform 분포로 사용
            'lambda_l1': trial.suggest_loguniform('lambda_l1', 1e-8, 10.0),
            # L2 값을 1e-8-10.0까지 로그 uniform 분포로 사용
            'lambda_l2': trial.suggest_loguniform('lambda_l2', 1e-8, 10.0),
            'seed': SEED,
            'verbose': -1,
            'n_jobs': -1,    
        }
    elif model_name == 'catboost':
        model_params = {
            'n_estimators': 10000,
            'learning_rate' : trial.suggest_loguniform('learning_rate', 0.01, 0.3),
            'eval_metric': 'AUC', # 평가 지표 설정
            'loss_function': 'Logloss', # 손실 함수 설정
            'random_seed': SEED,
            'metric_period': 100,
            'od_wait': 100, # early stopping round
            'depth' : trial.suggest_int('depth', 4, 10),
            'rsm': trial.suggest_uniform('rsm', 0.4, 1.0)        
        }
    
    elif model_name == 'xgb':
        model_params = {        
            'objective': 'binary:logistic', 
            'learning_rate' : trial.suggest_loguniform('learning_rate', 0.01, 0.3),
            'max_depth': trial.suggest_int("max_depth", 3, 9),        
            'lambda': trial.suggest_loguniform('lambda', 1e-8, 1.0),
            'alpha': trial.suggest_loguniform('alpha', 1e-8, 1.0),
            "subsample": trial.suggest_float("subsample", 0.2, 1.0),
            "colsample_bytree": trial.suggest_float("colsample_bytree", 0.2, 1.0),
            "min_child_weight": trial.suggest_int("min_child_weight", 2, 10)
        }

    # oof prediction 함수 호출해서 out of fold validation 예측값을 얻어옴
    y_oof, test_preds, fi = make_oof_prediction(train, y, test, features, model_params=model_params,folds = 5)
    
    # Validation 스코어 계산
    val_auc = roc_auc_score(label, y_oof)
    
    return val_auc
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=10) # 10회 동안 하이퍼 파라미터 탐색
```

```python
def best_params_model(best_params,model_name = model_name):
    if model_name == 'lightgbm':
        model_params = {
            'objective': 'binary', # 이진 분류
            'boosting_type': 'gbdt',
            'num_leaves': best_params['num_leaves'],
            'max_bin': best_params['max_bin'],             
            'min_data_in_leaf':best_params['min_data_in_leaf'],            
            'feature_fraction': best_params['feature_fraction'],            
            'bagging_fraction': best_params['bagging_fraction'],            
            'bagging_freq': best_params['bagging_freq'],
            'n_estimators': 10000, # 트리 개수
            'early_stopping_rounds': 100,            
            'lambda_l1': best_params['lambda_l1'],           
            'lambda_l2': best_params['lambda_l2'],
            'seed': SEED,
            'verbose': -1,
            'n_jobs': -1,    
        }
    elif model_name == 'catboost':
        model_params = {
            'n_estimators': 10000,
            'learning_rate' : best_params['learning_rate'],
            'eval_metric': 'AUC', # 평가 지표 설정
            'loss_function': 'Logloss', # 손실 함수 설정
            'random_seed': SEED,
            'metric_period': 100,
            'od_wait': 100, # early stopping round
            'depth' : best_params['depth'],
            'rsm': best_params['rsm']        
        }
    
    elif model_name == 'xgb':
        model_params = {        
            'objective': 'binary:logistic', 
            'learning_rate' : best_params['learning_rate'],
            'max_depth': best_params['max_depth'],        
            'lambda': best_params['lambda'],
            'alpha': best_params['alpha'],
            "subsample": best_params['subsample'],
            "colsample_bytree": best_params['colsample_bytree'],
            "min_child_weight": best_params['min_child_weight']
        }
    
    return model_params
```

# plot_feature_importances

```python
def plot_feature_importances(df, n=20, color='blue', figsize=(12,8)):
    # 피처 중요도 순으로 내림차순 정렬
    df = df.sort_values('importance', ascending = False).reset_index(drop = True)
    
    # 피처 중요도 정규화 및 누적 중요도 계산
    df['importance_normalized'] = df['importance'] / df['importance'].sum()
    df['cumulative_importance'] = np.cumsum(df['importance_normalized'])
    
    plt.rcParams['font.size'] = 12
    plt.style.use('fivethirtyeight')
    # 피처 중요도 순으로 n개까지 바플롯으로 그리기
    df.loc[:n, :].plot.barh(y='importance_normalized', 
                            x='feature', color=color, 
                            edgecolor='k', figsize=figsize,
                            legend=False)

    plt.xlabel('Normalized Importance', size=18); plt.ylabel(''); 
    plt.title(f'Top {n} Most Important Features', size=18)
    plt.gca().invert_yaxis()
    
    return df
```
